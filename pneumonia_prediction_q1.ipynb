Q1.1

Below we plot images of a normal case, a bacterial pneumonia case, and a viral pneumonia case.

Regarding the label distribution:
- Train: 1341 Healthy - 3875 Pneumonia
- Test:  234  Healthy -  390 Pneumonia
- Val:     8  Healthy -    8 Pneumonia

```python
img_normal1 = np.asarray(Image.open('data_part2/train/NORMAL/IM-0115-0001.jpeg'))
img_pneumonia_bacterial1 = np.asarray(Image.open('data_part2/train/PNEUMONIA/person1_bacteria_1.jpeg'))
img_pneumonia_viral1 = np.asarray(Image.open('data_part2/train/PNEUMONIA/person80_virus_150.jpeg'))

img_normal2 = np.asarray(Image.open('data_part2/train/NORMAL/IM-0203-0001.jpeg'))
img_pneumonia_bacterial2 = np.asarray(Image.open('data_part2/train/PNEUMONIA/person38_bacteria_190.jpeg'))
img_pneumonia_viral2 = np.asarray(Image.open('data_part2/train/PNEUMONIA/person967_virus_1640.jpeg'))


fig, axes = plt.subplots(2, 3, figsize=(10, 7))

axes[0, 0].imshow(img_normal1, cmap='gray')
axes[0, 0].set_title('Normal')

axes[0, 1].imshow(img_pneumonia_bacterial1, cmap='gray')
axes[0, 1].set_title('Pneumonia Bacterial')

axes[0, 2].imshow(img_pneumonia_viral1, cmap='gray')
axes[0, 2].set_title('Pneumonia Viral')

axes[1, 0].imshow(img_normal2, cmap='gray')
axes[1, 0].set_title('Normal')

axes[1, 1].imshow(img_pneumonia_bacterial2, cmap='gray')
axes[1, 1].set_title('Pneumonia Bacterial')

axes[1, 2].imshow(img_pneumonia_viral2, cmap='gray')
axes[1, 2].set_title('Pneumonia Viral')

plt.show()
```

Q1.3

One potential source of bias is that the x-rays may have been taken using different machines in different hospitals. Meaning that the model could pick up on the differences due to the different machines and not due to the patient being healthy or sick.

Q1.4

According to papers it was found that the performance of image classification on x-ray images was the best when using adaptive masking + histogram equalization + gaussian blur (see https://pmc.ncbi.nlm.nih.gov/articles/PMC8982897/). Howver, one might note that the difference is at most 7% compared to performing no pre-processing.
